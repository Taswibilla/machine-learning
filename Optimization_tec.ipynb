```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import SGD, RMSprop, Adam, AdamW

# Data
X = np.array([[1], [2], [3], [4], [5]], dtype=np.float32)
y = np.array([[2], [4], [6], [8], [10]], dtype=np.float32)

def create_model():
    model = Sequential([
        Input(shape=(1,)),
        Dense(8, activation='relu'),
        Dense(1)
    ])
    return model

optimizers = {
    "SGD": SGD(learning_rate=0.01),
    "Momentum": SGD(learning_rate=0.01, momentum=0.9),
    "RMSprop": RMSprop(learning_rate=0.001),
    "Adam": Adam(learning_rate=0.001),
    "AdamW": AdamW(learning_rate=0.001, weight_decay=0.0001)
}

for name, opt in optimizers.items():

    model = create_model()

    model.compile(
        optimizer=opt,
        loss='mse'
    )

    model.fit(X, y, epochs=50, verbose=0)

    prediction = model.predict(
        np.array([[6]], dtype=np.float32),
        verbose=0
    )

    print(f"{name} Prediction for x=6 : {prediction[0][0]:.2f}")

```
