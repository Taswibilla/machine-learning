```python
docs = [
    "Students must maintain 75% attendance.",
    "Semester exams are conducted twice a year.",
    "Library timing is 9 AM to 5 PM."
]

print(docs)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(docs)

print(embeddings.shape)
```

```python
import faiss

index = faiss.IndexFlatL2(embeddings.shape[1])

index.add(embeddings)

print("Documents Added")
```

```python
query = "Students must maintain 75% attendance?"

query_embedding = model.encode([query])

D, I = index.search(query_embedding, k=1)

print("Matched Document:")
print(docs[I[0][0]])
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
query = "What is the attendance requirement?"

query_embedding = model.encode([query])

D, I = index.search(query_embedding, k=1)

context = docs[I[0][0]]

print(context)


prompt = f"""
Use the context below to answer the question.

Context:
{context}

Question:
{query}

Answer:
"""

result = generator(
    prompt,
    max_new_tokens=50
)

print(result[0]["generated_text"])
```
