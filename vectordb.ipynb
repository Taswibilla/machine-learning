```python
import numpy as np

document = """
Machine Learning is a subset of Artificial Intelligence.
Deep Learning uses neural networks.
Cricket is a popular sport.
Football is played worldwide.
"""

chunks = [line.strip() for line in document.split("\n") if line.strip()]

print("Chunks:")
for chunk in chunks:
    print("-", chunk)

# Dummy embedding for learning purposes
# Real systems use OpenAI, BGE, E5 etc.

def embed(text):
    text = text.lower()

    ai_words = ["machine", "learning", "ai", "artificial", "neural"]

    sport_words = ["cricket", "football", "sport"]

    ai_score = sum(word in text for word in ai_words)
    sport_score = sum(word in text for word in sport_words)

    return np.array([ai_score, sport_score])

vector_db = []

for chunk in chunks:

    vector_db.append({
        "text": chunk,
        "embedding": embed(chunk)
    })

print("\nStored Vectors:")
for item in vector_db:
    print(item)

query = "What is Machine Learning?"

query_embedding = embed(query)

print("\nQuery Embedding:")
print(query_embedding)

def cosine_similarity(a, b):

    return np.dot(a, b) / (
        np.linalg.norm(a) *
        np.linalg.norm(b)
    )


results = []

for item in vector_db:

    score = cosine_similarity(
        query_embedding,
        item["embedding"]
    )

    results.append(
        (score, item["text"])
    )

results.sort(reverse=True)

top_k = 2

print("\nTop Results:")

for score, text in results[:top_k]:
    print(f"{score:.3f} -> {text}")
```
