```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input

# Dataset
X = np.array([[1], [2], [3], [4]], dtype=np.float32)
y = np.array([[2], [4], [6], [8]], dtype=np.float32)

# Model
model = Sequential([
    Input(shape=(1,)),
    Dense(1)
])

# Compile
model.compile(
    optimizer='sgd',
    loss='mse'
)

# Train.     when we run these automatically it processes the bp
model.fit(
    X,
    y,
    epochs=100,
    verbose=1
)

# Prediction
prediction = model.predict(np.array([[5]], dtype=np.float32))
print("Prediction:", prediction)
```
